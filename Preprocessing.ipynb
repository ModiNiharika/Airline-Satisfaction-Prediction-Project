# Data cleaning

### 1. Handling missing values

```python
df.isnull().sum()
```

### 2. Checking the Null value rows

```python
df[df["Arrival Delay in Minutes"].isnull()]
```

```python
df["Arrival Delay in Minutes"]=(df['Arrival Delay in Minutes'].astype(str).str.strip())
df["Arrival Delay in Minutes"]=pd.to_numeric(df["Arrival Delay in Minutes"], errors="coerce")
df.isnull().sum()
```

> We removed hidden/extra whitespace from the column and standardized it to numeric. After this step, 83 values were still truly missing, so we handled them separately.

```python
df["Arrival Delay in Minutes"]=df["Arrival Delay in Minutes"].fillna(df["Arrival Delay in Minutes"].median())
df.isnull().sum()
```

> We used Median Imputation, because it prevents distortion in Model Prediction.
After this step, there are no missing values remaining in the dataset.

# Data Formatting

### 1. Converting Target column to Numbers

```python
df["Satisfaction"] = (df["Satisfaction"].astype(str).str.strip().str.lower())
```

```python
df["Satisfaction"] = df["Satisfaction"].map({
    "satisfied": 1,
    "neutral or dissatisfied": 0
})

```

```python
display(df)
```

### 2. Removing the duplicate rows

```python
rows_before = df.shape[0]
df.drop_duplicates(inplace=True)
print(f"Removed {rows_before - df.shape[0]} duplicate rows. Remaining rows: {df.shape[0]}")
```

# Binning

> Binning is applied to the Age column to categorize ages into different groups.

```python
df["Age_group"]=pd.cut(df["Age"], bins=[0,18, 30, 45, 60, 100],
                        labels=["Teen", "Young", "Mid", "Senior", "Old" ])
```

```python
df.insert(df.columns.get_loc("Age")+1, "Age_group", df.pop("Age_group"))
```

# Grouping + ANOVA

> Grouping is performed to divide the data into different categories based on travel class which helps in comparing customer satisfaction level across different groups.

```python
df["Class"].unique()
```

```python
import scipy.stats as stats
group1=df[df["Class"]=="Business"]["Satisfaction"]
group2=df[df["Class"]=="Eco"]["Satisfaction"]
group3=df[df["Class"]=="Eco Plus"]["Satisfaction"]
print(stats.f_oneway(group1, group2, group3))
```
